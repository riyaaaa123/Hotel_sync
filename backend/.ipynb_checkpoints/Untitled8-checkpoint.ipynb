```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from xgboost import XGBRegressor
from sklearn.base import BaseEstimator, TransformerMixin
from mlmodel.ml_utils import HolidayFeatureEngineer


# ----------------------------
# Load Data
# ----------------------------
df = pd.read_csv("hotel_bookings_india_updated_with_regional.csv", parse_dates=['date'])

df['booking_rate'] = df['number_of_bookings'] / df['total_rooms']
df['price_discount_interaction'] = df['room_price'] * df['discount_rate']
df['price_per_room'] = df['room_price'] / (df['total_rooms'] + 1)
df['booking_cancellation_ratio'] = df['number_of_bookings'] / (df['number_of_cancellations'] + 1)
df.dropna(inplace=True)
df_sampled = df.sample(n=15000, random_state=42)

# ----------------------------
# Features & Targets
# ----------------------------
features = ['date', 'city', 'star_rating', 'total_rooms', 'room_price',
            'discount_rate', 'is_refundable', 'season',
            'booking_rate', 'price_discount_interaction', 'price_per_room',
            'booking_cancellation_ratio']

X = df_sampled[features]
y_bookings = df_sampled['number_of_bookings']
y_cancellations = np.log1p(df_sampled['number_of_cancellations'])

X_train, X_test, yb_train, yb_test, yc_train, yc_test = train_test_split(
    X, y_bookings, y_cancellations, test_size=0.2, random_state=42
)

# Add light noise
np.random.seed(42)
yb_train_noisy = yb_train + np.random.normal(0, 0.1, size=yb_train.shape)
yc_train_noisy = yc_train + np.random.normal(0, 0.01, size=yc_train.shape)

# ----------------------------
# Pipelines
# ----------------------------
categorical_features = ['city', 'season', 'region']
preprocessor = ColumnTransformer([
    ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
], remainder='passthrough')

bookings_pipeline = Pipeline([
    ('holiday_features', HolidayFeatureEngineer()),
    ('preprocessor', preprocessor),
    ('regressor', XGBRegressor(
        n_estimators=20, max_depth=4, learning_rate=1,
        objective='reg:squarederror', random_state=42
    ))
])

cancellations_pipeline = Pipeline([
    ('holiday_features', HolidayFeatureEngineer()),
    ('preprocessor', preprocessor),
    ('regressor', XGBRegressor(
        n_estimators=40, max_depth=4, learning_rate=1,
        subsample=0.9, colsample_bytree=0.85,
        objective='reg:squarederror', random_state=42
    ))
])

# ----------------------------
# Train & Evaluate
# ----------------------------
bookings_pipeline.fit(X_train, yb_train_noisy)
cancellations_pipeline.fit(X_train, yc_train_noisy)

yb_pred = bookings_pipeline.predict(X_test)
yc_pred_log = cancellations_pipeline.predict(X_test)
yc_pred = np.expm1(yc_pred_log)
yc_test_actual = np.expm1(yc_test)

print("\n--- ✅ Final Model Results (Balanced) ---")
print(f"MAE - Bookings: {mean_absolute_error(yb_test, yb_pred):.2f}")
print(f"MAE - Cancellations: {mean_absolute_error(yc_test_actual, yc_pred):.2f}")
print(f"RMSE - Bookings: {np.sqrt(mean_squared_error(yb_test, yb_pred)):.2f}")
print(f"RMSE - Cancellations: {np.sqrt(mean_squared_error(yc_test_actual, yc_pred)):.2f}")
print(f"R² - Bookings: {r2_score(yb_test, yb_pred):.4f}")
print(f"R² - Cancellations: {r2_score(yc_test_actual, yc_pred):.4f}")
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Get feature names from preprocessor
feature_names = bookings_pipeline.named_steps['preprocessor'].get_feature_names_out()

# Get feature importances from trained model
importances = bookings_pipeline.named_steps['regressor'].feature_importances_

# Plot top 15
importance_df = pd.DataFrame({'feature': feature_names, 'importance': importances})
importance_df = importance_df.sort_values(by='importance', ascending=False)

plt.figure(figsize=(10, 6))
plt.barh(importance_df['feature'][:15][::-1], importance_df['importance'][:15][::-1])
plt.title('Top 15 Feature Importances - Bookings Model')
plt.xlabel('Importance')
plt.tight_layout()
plt.show()
```

```python
residuals = yb_test - yb_pred

plt.figure(figsize=(8, 5))
plt.hist(residuals, bins=40, edgecolor='black')
plt.title("Residual Distribution - Bookings")
plt.xlabel("Prediction Error (Actual - Predicted)")
plt.grid(True)
plt.show()
```
